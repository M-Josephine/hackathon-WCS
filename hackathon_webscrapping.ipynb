```
url_pages = ['https://www.marmiton.org/recettes/recherche.aspx?aqt=saint-valentin&page=1',]
i = 0

for page in range(2,57):
    i = i+12
    url = "https://www.marmiton.org/recettes/recherche.aspx?aqt=saint-valentin&start={}&page={}".format(i,str(page))
    url_pages.append(url)

url_pages
```

```
liste_url_globale = []

for url in url_pages:
    html = requests.get(url, headers= {'User-Agent': navigator})
    soup = BeautifulSoup(html.text, 'html.parser')

    url_receipts = soup.find_all('div', {"class" : "recipe-card"})

    for div in url_receipts:
        liste_url_globale.append('https://www.marmiton.org' + div.find('a', {'class':'recipe-card-link'})['href'])
        
```

```
import time
```

```
liste_titre =[]
liste_niveau =[]
liste_budget =[]
liste_prep=[]
liste_cuisson=[]
liste_note=[]
liste_tags=[]
liste_sauvegarde=[]
liste_partage=[]
liste_etape=[]
liste_ingredients = []
liste_nb_ingredients=[]
liste_nb_ustensiles=[]
liste_nb_com=[]
liste_url = []

#liste_url_short=liste_url_globale[1:4] #short liste pour essai
#liste_url_globale=liste_url_short

for url in liste_url_globale:
    liste_url.append(url)
    #Connexion site rcupration page
    html = requests.get(url, headers= {'User-Agent': navigator})
    soup = BeautifulSoup(html.text, 'html.parser')
    #titre
    titre = soup.find('h1', {'class' : "main-title"}).text
    liste_titre.append(titre.strip())
    #niveau difficult
    niveau =soup.find_all('div', {'class' : "recipe-infos__level"})
    if len(niveau) >0:
        niveau = niveau[0].find('span', {'class' : "recipe-infos__item-title"}).text
        liste_niveau.append(niveau.strip())
    else:
        liste_niveau.append('vide')
    #Nbre de commentaires
    nb_com =soup.find_all('div', {'class' : "recipe-infos-users__item recipe-infos-users__rate"})

    if len(nb_com)>0:
        nb_com = int(nb_com[0].find('span', {'class' : "recipe-infos-users__value"}).text)
        liste_nb_com.append(nb_com)
    else:
        liste_nb_com.append(0)
    #Niveau budget
    budget =soup.find_all('div', {'class' : "recipe-infos__budget"})
    if len(budget) >0:
        budget = budget[0].find('span', {'class' : "recipe-infos__item-title"}).text
        liste_budget.append(budget.strip())
    else:
        liste_budget.append('vide')
    #temps prparation
    prep=soup.find_all('div', {'class' : "recipe-infos__timmings__preparation"})
    if len(prep) >0:
        liste_prep.append(prep[0].text.replace('Prparation\xa0: ','').strip())
    else:
        liste_prep.append('0 min')

    #temps cuisson
    cuisson=soup.find_all('div', {'class' : "recipe-infos__timmings__cooking"})
    if len(cuisson) >0:
        cuisson = cuisson[0].find('span', {'class' : "recipe-infos__timmings__value"}).text
        liste_cuisson.append(cuisson.strip())
    else:
        liste_cuisson.append('0 min')

    #Note moyenne
    note=soup.find('span', {'class' : "recipe-infos-users__rating"}).text
    liste_note.append(note.strip())

    #Tags
    tags=soup.find('ul', {'class' : "mrtn-tags-list"}).text
    tags = tags.rstrip('\n')
    tags = tags.lstrip('\n')
    tags = tags.split('\n')
    liste_tags.append(tags)

    #Nbre de sauvegarde (favoris)
    sauvegarde=soup.find_all('div', {'class' : "recipe-infos-users__notebook"})
    sauvegarde=sauvegarde[0].find('span', {'class' : "recipe-infos-users__value"}).text
    liste_sauvegarde.append(sauvegarde.strip())

    #Nbre de partage
    partage=soup.find_all('div', {'class' : "recipe-infos-users__item recipe-infos-users__share"})
    partage=partage[0].find('span', {'class' : "recipe-infos-users__value"}).text
    liste_partage.append(partage.strip())

    #Nbre tape
    etape=len(soup.find_all('li', {'class' : "recipe-preparation__list__item"}))
    liste_etape.append(etape)

    #Liste ingrdients
    ingredients=soup.find_all('li', {'class' : "recipe-ingredients__list__item"})
    list_ingredients = []
    for i in ingredients:
        ing = i.find('span',{'class': 'ingredient'})
        list_ingredients.append(ing.text)
        
    liste_ingredients.append(list_ingredients)

    #Nbre ingrdients
    if len(liste_ingredients)>0: #peu probable qu'il n'y ait pas d'ingrdients mais bon...
        liste_nb_ingredients.append(len(list_ingredients))
    else:
        liste_nb_ingredients.append(0)
    #Nbre ustensile
    ustensile=soup.find_all('li', {"class":"recipe-utensils__list__item"})
    if len(ustensile)>0:
        liste_nb_ustensiles.append(len(ustensile))
    else:
        liste_nb_ustensiles.append(0)

    time.sleep(2)
```

```
print(len(liste_titre)) 
print(len(liste_niveau))
print(len(liste_nb_com))
print(len(liste_budget))
print(len(liste_prep))
print(len(liste_cuisson))
print(len(liste_note))
print(len(liste_tags))
print(len(liste_sauvegarde))
print(len(liste_partage))
print(len(liste_etape))
print(len(liste_ingredients))
print(len(liste_url)) 
```

```
df = pd.DataFrame({'titre': liste_titre, 'niveau_difficulte':liste_niveau,
'Nbre Commentaires' : liste_nb_com, 'budget' : liste_budget,'temps_preparation':liste_prep,"Nbre Ustensiles":liste_nb_ustensiles,"Nbre Ingrdients":liste_nb_ingredients,
'temps_cuisson': liste_cuisson, 'note':liste_note, 'tags':liste_tags, 'nb_sauvegarde':liste_sauvegarde, 'nb_partage': liste_partage, 'nb_etapes':liste_etape,
'ingredients' : liste_ingredients, 'url_recette':liste_url})

df.head()
```
